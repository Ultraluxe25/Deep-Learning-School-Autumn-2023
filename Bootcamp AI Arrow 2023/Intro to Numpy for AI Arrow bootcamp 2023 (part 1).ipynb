```python
from PIL import Image
import numpy as np
```

```python
cat_image = 'cat.jpg'
with Image.open(cat_image) as cat:
    cat.load()
```

```python
cat
```

```python
# Picture size
width, height = cat.size
width, height
```

```python
# Matrix from picture
img = np.array(cat)
```

```python
# Get filter
filt = np.copy(cat)
filt[..., 0] = 1
filt
```

```python
img_cat_filt = Image.fromarray(obj=filt)
img_cat_filt
```

```python
img255 = np.copy(cat)
img255[..., 0] = 255
img255[..., 2] = 1
img_255 = Image.fromarray(img255)
img_255
```
