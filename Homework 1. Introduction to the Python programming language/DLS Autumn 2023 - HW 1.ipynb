## Задача 1
### Создайте программу, которая три раза выводит фразу “какой чудесный день” c новой строки.

Ожидаемый результат работы программы:

какой чудесный день

какой чудесный день

какой чудесный день

**Sample Input:**

**Sample Output:**

какой чудесный день
какой чудесный день
какой чудесный день

```python
def say(text: str) -> None:
    for _ in range(3):
        print(text)
        
        
response = 'какой чудесный день'
say(response)
```

## Задача 2
### Считайте имя пользователя и выведите на экран фразу “Как дела, <username>?”, где вместо <username> должно стоять считанное значение.

**Sample Input:**

1234
    
**Sample Output:**

Как дела, 1234?

```python
class Response():
    def __init__(self, name):
        self.name = name
        
    def response(self):
        return f'Как дела, {self.name}?'
    
    
user_name = input()
answer = Response(user_name)
print(answer.response())
```

## Задача 4
### Напишите программу "калькулятор", которая считает два натуральных числа a и b
1 ⩽ a, b ⩽ 10 и вывести 5 строк, в которых будут напечатаны следующие числа:

a + b,

a − b,

a ⋅ b,

[a / b] (целочисленное деление),

a ** b.

**Sample Input:**

1 2

**Sample Output:**

3

-1

2

0

1

```python
class Calculator:
    def __init__(self, x: int, y: int) -> None:
        self.x = x
        self.y = y
        
    def addition(self):
        return self.x + self.y
    
    def subtraction(self):
        return self.x - self.y
    
    def multiplication(self):
        return self.x * self.y
    
    def integer_division(self):
        return self.x // self.y
    
    def exponentation(self):
        return self.x ** self.y
    
    
a, b = map(int, input().split())
my_calculator = Calculator(a, b)

add = my_calculator.addition()
subtract = my_calculator.subtraction()
multiplication = my_calculator.multiplication()
int_div = my_calculator.integer_division()
exp = my_calculator.exponentation()

operations = (add, subtract, multiplication, int_div, exp)
for activity in operations:
    print(activity)
```

## Задача 5
### Напишите программу, которая принимает на вход два целых числа 

1⩽a,b⩽100 и выводит на экран сумму двух чисел:

* остатка от деления первого числа на второе;

* результата целочисленного деления второго числа на первое.

**Sample Input:**

12

7

**Sample Output:**

5

```python
a, b = [int(input()) for _ in range(2)]
number_1 = a % b
number_2 = b // a
print(number_1 + number_2)
```
